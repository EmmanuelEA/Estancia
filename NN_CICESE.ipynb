```python
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression 
import statsmodels.formula.api as smf
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_model,svm
from sklearn.svm import SVR
from time import time
from sklearn.model_selection import GridSearchCV
from joblib import dump, load 
from keras.models import Sequential
from keras.layers.core import Dense
```

```python
datamain = pd.read_csv('CICESEDATA.csv')
datamain.head()

datamain = datamain.fillna(0)
datamain.describe()
feature_cols = ["Mes", "Dia", "Anio", "Evaporacion", "Calor", "Precipitacion"]
y_col = ['TempMax']

df1922to2015 = datamain[(datamain['Anio']>=1922) & (datamain['Anio'] <= 2015)]
df2016 = datamain[(datamain['Anio'] >= 2016)]
dXtrain, dYtrain = df1922to2015[feature_cols] ,  df1922to2015[y_col]
dXtest, dYtest = df2016[feature_cols] ,  df2016[y_col]


X = dXtrain
Y = dYtrain['TempMax']
X.shape, Y.shape
```

```python
# cargamos datos
training_data = np.array(dXtrain, "float32")
testing_data = np.array(dXtest, "float32")

# y estos son los resultados que se obtienen, en el mismo orden
target_data = np.array(dYtrain, "float32")
test_target_data = np.array(dYtest, "float32")
```

```python
model = Sequential()
model.add(Dense(16, input_dim=6, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
 
model.compile(loss='mean_squared_error',
              optimizer='adam',
              metrics=['binary_accuracy'])
 
```

```python
model.fit(training_data, target_data, epochs=200)
```

```python
# evaluamos el modelo
scores = model.evaluate(training_data, target_data)
 
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
print (model.predict(training_data).round())
```

```python
y = model.predict(testing_data)
y
```

```python
start_time = time()
svr_rbf = SVR(kernel="rbf", C=2000, gamma=0.01)
y_rbf = svr_rbf.fit(X,Y)
dump(y_rbf, 'SVMC2000G01.joblib')
yrbf = y_rbf.predict(dXtrain)
elapsed_time = (time()-start_time)/60
elapsed_time
SSD = np.sum((Y - yrbf)**2)
RSE = np.sqrt(SSD/(len(X)-1))
y_mean = np.mean(Y)
error = RSE/y_mean
print("Valor C:",2000)
print("Elapsed time: ",elapsed_time)
print("R2:" +str(svr_rbf.score(X, Y)))
print("SSD: "+str(SSD)+", RSE: " +str(RSE) + ", Y_mean: " +str(y_mean) +", error: " + str(error*100)+ "%")
```
