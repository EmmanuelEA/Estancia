# # Obtener datos con libreria urllib3

```python
import urllib3
import csv
```

```python
medals_url = "http://winterolympicsmedals.com/medals.csv"
```

```python
medals_data = pd.read_csv(medals_url)
```

```python
medals_data.head()
```

```python
#se obtiene los datos y se codifican
http = urllib3.PoolManager()
r = http.request('GET', medals_url)
response = r.data.decode('UTF-8')
response
```

```python
#Separamos el archivo por los saltos de linea
datos = response.split("\n")
datos
```

```python
#obtenemos las columnas que son de la primer linea
cols = datos[0].split(',')
n_cols = len(cols)
n_cols
```

```python
#Generamos un diccionario vacÃ­o donde irÃ¡ la informaciÃ³n procesada desde la URL externa
counter = 0
main_dict = {}
for col in cols:
    main_dict[col] = []
main_dict
```

```python
#Procesamos fila a fila la informaciÃ³n para ir rellenando el diccionario con los datos como hicimos antes
for line in datos:
    #Nos saltamos la primera lÃ­nea que es la que contiene la cabecera y ya tenemos procesada
    if(counter > 0):
        #Dividimos cada string por las comas como elemento separador
        values = line.strip().split(',')
        #AÃ±adimos cada valor a su respectiva columna del diccionario
        for i in range(len(cols)):
            main_dict[cols[i]].append(values[i])
    counter += 1

print("El data set tiene %d filas y %d columnas"%(counter, n_cols))
```

```python
main_dict
```

```python
#Creamos el dataframe con pandas desde el diccionario creado
dfmedals = pd.DataFrame(main_dict)
dfmedals.head()
```
